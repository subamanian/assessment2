```python
import pandas as pd
```

```python
df=pd.read_csv('CaVideos.csv')
df.head()
```

```python
df.drop(df.columns[[6,11,12,13,13,14,15]],axis=1,inplace=True)
df
```

```python
sorted_df=df.sort_values(["views"],ascending=False)
sorted_df.head(20)
```

```python
sorted_df.rename(columns={"publish_time":"publis_datetime","comment_count":"comments"},inplace=True)
sorted_df.head()
```

```python
asorted_df=df.sort_values(["views"],ascending=False)
asorted_df.head(20)
```

```python
df1=pd.read_csv('UsVideos.csv')
df1.head()
```

```python
df1.drop(df1.columns[[6,11,12,13,13,14,15]],axis=1,inplace=True)
df1
```

```python
sorted1_df=df1.sort_values(["views"],ascending=False)
sorted1_df.head(20)
```

```python
sorted1_df.rename(columns={"publish_time":"publis_datetime","comment_count":"comments"},inplace=True)
sorted1_df.head()
```

```python
asorted1_df=df1.sort_values(["views"],ascending=False)
asorted1_df.head(20)
```

```python
category="US_category_id.json"
category_df=pd.read_json(category)
category_df.head()
```

```python
category_df["items"][0]
```

```python
making_into_list=category_df["items"].tolist()
```

```python
category_id=[]
category_title=[]
for item in making_into_list:
    category_id.append(int(item["id"]))
    category_title.append(item["snippet"]["title"])
```

```python
new_category_df=pd.DataFrame({"category_title":category_title,"category_id":category_id})
```

```python
new_category_df
```

```python
CA_df=pd.merge(sorted_df,new_category_df,on="category_id",how="inner")
CA_df.head()
```

```python
CA_df['country']='CA'
CA_df
```

```python
CA1_df=pd.merge(asorted_df,new_category_df,on="category_id",how="inner")
CA1_df.head()
```

```python
CA1_df['country']='CA'
CA1_df
```

```python
Us_df=pd.merge(sorted1_df,new_category_df,on="category_id",how="inner")
Us_df.head()
```

```python
Us1_df=pd.merge(asorted1_df,new_category_df,on="category_id",how="inner")
Us1_df.head()
```

```python
Us1_df['country']='US'
Us1_df
```
